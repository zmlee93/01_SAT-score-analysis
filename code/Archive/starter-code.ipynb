## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:

def man_stdev(num_list):
    first_term = 1 / len(num_list)
    avg = man_mean(num_list)
    second_term = 0
    for index in range(len(num_list)):
        second_term += (num_list[index] - avg)**2
        
    return (first_term * second_term)**0.5

num_list = [1,2,3,4,5]
man_stdev(num_list)

# # #debug
# import statistics as stats
# stats.pstdev(num_list)
```

# Identify states with poorer performance post-covid

    To determine if a state has performed poorer due to covid lockdowns, a baseline standard should be established.
    In this case, the average of the results from 2017 to 2019 shall be used as the baseline.
    
    For a state to classify as having performance affected negatively, 
        the lowest score from 2020 and 2021 must be lower than the poorest performance in the 2017-2019 period.
        
    If a state is flagged out to be affected, it is worth considering comissioning another project to identify the
    root causes of the poor performance.

#### Calculate minimum of 2017-2019 results

```python
### Calculate min of 2017-2019 results

state_min = sat_scores[(sat_scores['year'] == 2017) |(sat_scores['year'] == 2018)|(sat_scores['year'] == 2019)]
state_min = state_min.drop(columns = ['year'])
state_min = state_min.groupby(by = ['state', 'postal_of_states']).min()
state_min.reset_index(inplace = True)
state_min = state_min.melt(id_vars = ['state','postal_of_states'], var_name = 'category')
state_min.sort_values(by = ['state','postal_of_states', 'category'], ascending = True, inplace = True, ignore_index = True)
state_min.rename(columns = {'value': '2017_2019_min'}, inplace = True)

#debug
# state_min.head(3)
```

#### Join with 2020 and 2021 data

```python
#prepare sat_2020 for merge 
sat_2020_unpivot = sat_2020.drop(columns = ['year'])
sat_2020_unpivot = sat_2020_unpivot.melt(id_vars = ['state'], var_name = 'category')
sat_2020_unpivot.sort_values(by = ['state', 'category'], ascending = True, inplace = True, ignore_index = True)
sat_2020_unpivot.rename(columns = {'value': '2020'}, inplace = True)

#prepare sat_2021 for merge 
sat_2021_unpivot = sat_2021.drop(columns = ['year'])
sat_2021_unpivot = sat_2021_unpivot.melt(id_vars = ['state'], var_name = 'category')
sat_2021_unpivot.sort_values(by = ['state', 'category'], ascending = True, inplace = True, ignore_index = True)
sat_2021_unpivot.rename(columns = {'value': '2021'}, inplace = True)

#join with average scores
state_min_w_2020_2021 = pd.merge(left = state_min,
                           right = sat_2020_unpivot,
                           how = 'inner',
                           on = ['state', 'category'])

state_min_w_2020_2021 = pd.merge(left = state_min_w_2020_2021,
                           right = sat_2021_unpivot,
                           how = 'inner',
                           on = ['state', 'category'])

# #debug
# state_min_w_2020_2021.head(3)
```

#### Create new column with name = '2020_2021_min', and drop 2020 and 2021 columns. Also, drop rows where category == 'total'.

```python
state_poor_performer = state_min_w_2020_2021

state_poor_performer['2020_2021_min'] = state_poor_performer[['2020','2021']].min(axis = 1)
state_poor_performer = state_poor_performer.drop(columns = ['2020', '2021'])

# #debug
# state_poor_performer.head(3)
```

#### Create flag column to identify poor performing states. 

```python
state_poor_performer['is_poor_performer'] = np.where(state_poor_performer['2020_2021_min'] < state_poor_performer['2017_2019_min'], True, False )
state_poor_performer.head(3)
```

```python
state_poor_performer_total = state_poor_performer[(state_poor_performer['is_poor_performer'] == True) & (state_poor_performer['category'] == 'total')]
state_poor_performer_total['min_delta'] = state_poor_performer_total['2017_2019_min'] - state_poor_performer_total['2020_2021_min']
state_poor_performer_total = state_poor_performer_total.sort_values(by = 'min_delta', ascending = True)

state_poor_performer_math = state_poor_performer[(state_poor_performer['is_poor_performer'] == True) & (state_poor_performer['category'] == 'math')]
state_poor_performer_math['min_delta'] = state_poor_performer_math['2017_2019_min'] - state_poor_performer_math['2020_2021_min']
state_poor_performer_math = state_poor_performer_math.sort_values(by = 'min_delta', ascending = True)

state_poor_performer_reading = state_poor_performer[(state_poor_performer['is_poor_performer'] == True) & (state_poor_performer['category'] == 'reading')]
state_poor_performer_reading['min_delta'] = state_poor_performer_reading['2017_2019_min'] - state_poor_performer_reading['2020_2021_min']
state_poor_performer_reading = state_poor_performer_reading.sort_values(by = 'min_delta', ascending = True)
```
